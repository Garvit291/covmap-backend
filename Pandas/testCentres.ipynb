```python
import pandas
import requests
from pandas.io.json import json_normalize
```

```python
url='https://covid.icmr.org.in/index.php/testing-facilities?option=com_hotspots&view=jsonv4&task=gethotspots&hs-language=en-GB&page=1&per_page=5000&cat=&level=4&ne=43.782227%2C104.887112&sw=-3.761174%2C60.590237&c=21.957347%2C82.738675&fs=0&offset=0&format=raw'
```

```python
session = requests.Session()
```

```python
response=session.get(url)
```

```python
data=response.json()["items"]
```

```python
df=json_normalize(data)
```

### last id = 1577

```python
df=df[df["id"]>1567]
```

```python
df.rename(columns = {'lat':'latitude', 'lng':'longitude','cutDescription':'cutdescription'}, inplace = True)
```

```python
del df['multimedia']
del df['readmore']
del df['country']
del df['state']
```
